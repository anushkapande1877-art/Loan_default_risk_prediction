```python
# ================================
# Class Imbalance Analysis
# Purpose: Check distribution of default vs non-default
# Important for understanding modeling challenges
# ================================

default_counts = df['ProbDefault'].value_counts(normalize=True) * 100
print("Class Distribution (%):")
print(default_counts)

# Interpretation:
# If one class dominates heavily, model may need balancing techniques.
```

```python
# ================================
# Correlation with Target Variable
# Purpose: Identify strongest numerical drivers of default risk
# ================================

corr_matrix = df.corr(numeric_only=True)

# Sort correlation with ProbDefault
target_corr = corr_matrix['ProbDefault'].sort_values(ascending=False)

print("Top Correlated Features With Default:")
print(target_corr)

# Interpretation:
# Higher absolute correlation = stronger linear relationship with default risk
```

```python
# ================================
# Default Rate by Applicant Type
# Purpose: Understand which applicant category is riskier
# ================================

applicant_risk = df.groupby('ApplicantType')['ProbDefault'].mean().sort_values(ascending=False)

print("Default Rate by Applicant Type:")
print(applicant_risk)

# Higher mean = higher average default probability
```

```python
# ================================
# Default Rate by Employment Type
# Purpose: Identify high-risk employment categories
# ================================

employment_risk = df.groupby('EmploymentType')['ProbDefault'].mean().sort_values(ascending=False)

print("Default Rate by Employment Type:")
print(employment_risk)

# Helps in loan approval policy decisions
```

```python
# ================================
# Credit Score Segmentation
# Purpose: Create business-friendly risk groups
# ================================

df['CreditScoreGroup'] = pd.cut(
    df['CreditScore'],
    bins=[300, 580, 670, 740, 850],
    labels=['Poor', 'Fair', 'Good', 'Excellent']
)

credit_group_risk = df.groupby('CreditScoreGroup')['ProbDefault'].mean()

print("Default Rate by Credit Score Group:")
print(credit_group_risk)

# Business Insight:
# Lower credit groups should show higher default probability
```

```python
# ================================
# Skewness Analysis
# Purpose: Detect highly skewed features that may need log transformation
# ================================

skewness = df[['AnnualIncome', 'LoanAmount']].skew()
print("Skewness Values:")
print(skewness)

# If skewness > 1 â†’ highly right skewed
# Log transformation may improve modeling performance
```

```python
# ================================
# Outlier Detection using IQR
# Purpose: Identify extreme loan values that may distort models
# ================================

Q1 = df['LoanAmount'].quantile(0.25)
Q3 = df['LoanAmount'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['LoanAmount'] < lower_bound) | (df['LoanAmount'] > upper_bound)]

print("Number of LoanAmount Outliers:", len(outliers))

# Consider winsorization if extreme values are excessive
```

```python
# ================================
# Multicollinearity Check
# Purpose: Detect redundant features
# ================================

plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")

# If AnnualIncome and MonthlyIncome are highly correlated,
# one of them can be removed to avoid redundancy
```
