```python
# ================================
# Class Imbalance Analysis
# Purpose: Check distribution of default vs paid back
# Important for understanding modeling challenges
# ================================

default_counts = df['loan_paid_back'].value_counts(normalize=True) * 100
print("Class Distribution (%):")
print(default_counts)

# Interpretation:
# If one class dominates heavily, model may need balancing techniques.
```

```python
# ================================
# Correlation with Target Variable
# Purpose: Identify strongest numerical drivers of loan payback
# ================================

corr_matrix = df.corr(numeric_only=True)

# Sort correlation with loan_paid_back
target_corr = corr_matrix['loan_paid_back'].sort_values(ascending=False)

print("Top Correlated Features With Loan Paid Back:")
print(target_corr)

# Interpretation:
# Higher absolute correlation = stronger linear relationship with loan payback
```

```python
# ================================
# Loan Paid Back Rate by Education Level
# Purpose: Understand which education category is riskier
# ================================

education_risk = df.groupby('education_level')['loan_paid_back'].mean().sort_values(ascending=False)

print("Loan Paid Back Rate by Education Level:")
print(education_risk)

# Higher mean = higher average loan payback rate
```

```python
# ================================
# Loan Paid Back Rate by Employment Status
# Purpose: Identify high-risk employment categories
# ================================

employment_risk = df.groupby('employment_status')['loan_paid_back'].mean().sort_values(ascending=False)

print("Loan Paid Back Rate by Employment Status:")
print(employment_risk)

# Helps in loan approval policy decisions
```

```python
# ================================
# Credit Score Segmentation
# Purpose: Create business-friendly risk groups
# ================================

df['CreditScoreGroup'] = pd.cut(
    df['credit_score'],
    bins=[300, 580, 670, 740, 850],
    labels=['Poor', 'Fair', 'Good', 'Excellent']
)

credit_group_risk = df.groupby('CreditScoreGroup', observed=True)['loan_paid_back'].mean()

print("Loan Paid Back Rate by Credit Score Group:")
print(credit_group_risk)

# Business Insight:
# Lower credit groups should show lower loan payback probability
```

```python
# ================================
# Skewness Analysis
# Purpose: Detect highly skewed features that may need log transformation
# ================================

skewness = df[['annual_income', 'loan_amount']].skew()
print("Skewness Values:")
print(skewness)

# If skewness > 1 â†’ highly right skewed
# Log transformation may improve modeling performance
```

```python
# ================================
# Outlier Detection using IQR
# Purpose: Identify extreme loan values that may distort models
# ================================

Q1 = df['loan_amount'].quantile(0.25)
Q3 = df['loan_amount'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['loan_amount'] < lower_bound) | (df['loan_amount'] > upper_bound)]

print("Number of loan_amount Outliers:", len(outliers))

# Consider winsorization if extreme values are excessive
```
